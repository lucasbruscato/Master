```python
# import libraries
import PyPDF2
```

```python
# creating an object 
file = open('teste.pdf', 'rb')

# creating a pdf reader object
fileReader = PyPDF2.PdfFileReader(file)

wordIndex = -1
flagInAWord = 0
words = []

for i in range(fileReader.numPages):
    page = fileReader.getPage(i).extractText().lower()
    
    for j in range(len(page)):
        letter = page[j]
        
        if (not letter.isalpha()) and flagInAWord != 0:
            flagInAWord = 0
        elif letter.isalpha() and flagInAWord == 0:
            flagInAWord = 1
            wordIndex += 1
            words.append(letter)
        elif letter.isalpha() and flagInAWord != 0:
            words[wordIndex] += letter
```

```python
print(words)
```

```python
from collections import Counter

count = Counter(words)
```

```python
import csv

with open("output.csv", "wb") as f:
    writer = csv.writer(f)
    writer.writerows(count)
```
